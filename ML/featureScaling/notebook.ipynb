```python
import numpy as np
```

```python
data = np.array([
    [26, 50000],
    [29, 70000],
    [34, 55000], 
    [31, 41000]
])
```

Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
```

Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(data)
scaled_data
```
